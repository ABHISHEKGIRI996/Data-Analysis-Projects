## Derive the columns - Fail_Year, Fail_Month, Fail_Day based on Failure date field

```python
d1=bk2['Failure Date']
type(d1)
df=pd.DataFrame(d1)

df
```

```python
df.dtypes
```

```python
df['Failure Date']=pd.to_datetime(df['Failure Date'])
```

```python
df['Fail Year']=df['Failure Date'].dt.year
df['Fail Quarter']=df['Failure Date'].dt.quarter
df['Fail Month']=df['Failure Date'].dt.month
df['Fail Day']=df['Failure Date'].dt.day
df['Fail Week']=df['Failure Date'].dt.week
df['Fail Day of Week']=df['Failure Date'].dt.dayofweek
df
```

## Depict Fail_Month wise mean of Total Deposits and Total Assets on a bar chart, line chart and a scatter chart

```python

bk2['Fail Year']=df['Fail Year']
bk2['Fail Month']=df['Fail Month'].replace({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'})
bk2['Fail Day']=df['Fail Day']
bk2['Fail Quarter']=df['Fail Quarter']
bk2['Fail Week']=df['Fail Week']
```

```python
bk2
```

```python

x1=bk2.groupby('Fail Month')[['Total Deposits','Total Assets']].mean()
x1
```

```python


x1.plot(kind='bar',figsize=(8,8),title='Fail_Month wise mean of Total Deposits and Total Assets')
plt.ylabel('Mean')

plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.title('Fail_Month wise mean of Total Deposits and Total Assets')

plt.scatter(x=x1.index,y=x1['Total Deposits'],marker='d',color='red',s=100,label='Total Deposits')
plt.scatter(x=x1.index,y=x1['Total Assets'],marker='*',color='green',s=100,label='Total Assets')

plt.ylabel('Mean')
plt.xlabel('Fail Month')
plt.legend()
plt.show()
```

## Find min of Total Assets and mean of Estimated Loss based on Quarter and Insurance Fund where Total Deposits is over 50000 and Institution Type is Savings Bank. Depict the result on a bar chart and stacked bar chart (unstacking is not required) separately

```python
x7=bk2[['Total Assets','Estimated Loss','Fail Quarter','Insurance Fund','Total Deposits','Institution Type']][( (bk2['Total Deposits']>50000) & (bk2['Institution Type']=='SAVINGS BANK') )]

x7
```

```python
x8=x7.groupby(['Fail Quarter','Insurance Fund'])['Total Assets'].min()
print(x8)

x9=x7.groupby(['Fail Quarter','Insurance Fund'])['Estimated Loss'].mean()
print(x9)


x10=pd.merge(x8,x9,on=['Fail Quarter','Insurance Fund'],how='inner')

x10
```

```python

x10.plot(kind='bar',title='min of Total Assets and mean of Estimated Loss based on Quarter',figsize=(8,8))
plt.ylabel('MIN AND MAX')
plt.show()
```

```python
x10.plot(kind='bar',title='min of Total Assets and mean of Estimated Loss based on Quarter',figsize=(8,8),stacked=True)
plt.ylabel('MIN AND MAX')
plt.show()
```
