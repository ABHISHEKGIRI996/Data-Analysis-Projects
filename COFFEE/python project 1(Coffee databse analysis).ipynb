## Depict the quartile distribution of COGS where Sales is greater than 400

```python

# "Q0 quantile of arr : ", np.quantile(data, .0)) 
# "Q1 quantile of arr : ", np.quantile(data, .25)) 
# "Q2 quantile of arr : ", np.quantile(data, .50)) 
# "Q3 quantile of arr : ", np.quantile(data, .75)) 
# "100th quantile of arr : ", np.quantile(data, 1))


n7 = cf['COGS'][cf['Sales']>400]
n7
```

```python
Q0=np.quantile(n7,.0)
Q1=np.quantile(n7,.25)
Q2=np.quantile(n7,.50)
Q3=np.quantile(n7,.75)
Q4=np.quantile(n7,1)

qval=[Q0,Q1,Q2,Q3,Q4]
qind=['Q0','Q1','Q2','Q3','Q4']
```

```python
# plt.figure(figsize=(8,8))
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(10,8))
ax1.bar(qind,qval)
ax2.plot(qind,qval,marker='2',lw=2)
ax3.pie(qval,labels=qind,autopct="%0.1f%%")
ax4.scatter(qind,qval)
plt.show()
```
